#Leetcode in Python for the problem 0083: Remove Duplicates from Sorted List

## define
Given the $head$ of a sorted linked list, delete all duplicates such that each element appears only once. Return the linked list __sorted__ as well.

## Example 1:
list1.jpg

```
Input: head = [1,1,2]
Output: [1,2]
```

## Example 2:
list2.jpg

```
Input: head = [1,1,2,3,3]
Output: [1,2,3]
```

## leetcode sumit anser

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        current = head

        while current is not None and current.next is not None:
            if current.val == current.next.val:
                current.next = current.next.next
            else:
                current = current.next

        return head
````
